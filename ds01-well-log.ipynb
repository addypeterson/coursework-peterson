## Introduction

Well logging is a common practice in oil and gas industries, and consists of recording physical, chemical, and structural properties while drilling boreholes. In Expedition 372A, a Logging-While-Drilling (LWD) technique was used, meaning data is collected as the hole is drilled using tools integrated into the drill itself.

- **LWD Tools:** Gamma ray, resistivity, and ultrasonic caliper tools. 
- **Physical Measurement Tools:** Advanced Piston Corer Temperature tool, Temperature Dual-Pressure probe, Sediment Temperature Pressure tool.
- **Imaging Tools:** Infrared cameras (used in cores), smear slide analysis.


The collected data consists of a variety of sub-seafloor measurements, including porosity, resistivity, methane concentrations, and sediment composition. Both LWD and core analysis techniques were used. Data were presented in 1D formats for most analyses, and were plotted against depth. 2D and 3D data were used for seismic data, providing a "slice" or "cube" of the subsurface.

Data were gathered to test various hypotheses regarding the behaviors of gas hydrates and seafloor stability. Scientists wanted to identify the saturation, composition, and distribution of gas hydrates in the sediment, understand slow-moving landslides, and callibrate seismic data by applying recorded velocity logs. Better understanding of seafloor stability and gas hydrates contributes to improved tsunami and geohazard assessment and advancements in climate change research.
