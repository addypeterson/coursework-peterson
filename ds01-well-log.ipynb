## Introduction

Well logging is a common practice in oil and gas industries, and consists of recording physical, chemical, and structural properties while drilling boreholes. In Expedition 372A, a Logging-While-Drilling (LWD) technique was used, meaning data is collected as the hole is drilled using tools integrated into the drill itself.

- **LWD Tools:** Gamma ray, resistivity, and ultrasonic caliper tools. 
- **Physical Measurement Tools:** Advanced Piston Corer Temperature tool, Temperature Dual-Pressure probe, Sediment Temperature Pressure tool.
- **Imaging Tools:** Infrared cameras (used in cores), smear slide analysis.


The collected data consists of a variety of sub-seafloor measurements, including porosity, resistivity, methane concentrations, and sediment composition. Both LWD and core analysis techniques were used. Data were presented in 1D formats for most analyses, and were plotted against depth. 2D and 3D data were used for seismic data, providing a "slice" or "cube" of the subsurface.

Data were gathered to test various hypotheses regarding the behaviors of gas hydrates and seafloor stability. Scientists wanted to identify the saturation, composition, and distribution of gas hydrates in the sediment, understand slow-moving landslides, and callibrate seismic data by applying recorded velocity logs. Better understanding of seafloor stability and gas hydrates contributes to improved tsunami and geohazard assessment and advancements in climate change research.

## Interpretation

**Gamma Ray (NGR):** The gamma ray plot shows the measured levels of radioactivity in the formation, which gives information about the sediment. Higher radioactivity is usually detected in clays, while lovwer radioactivity occurs in larger sediments like sand or silt.

**Resistivity:** Resistivity is a measure that reveals how deep an electric current travels around sediments. Electric current prefers to travel through a conductor like saltwater, so since gas hydrate is an insulator, its presence between sediment particles increases resistivity. This allows for better imaging of the locations of gas hydrates.

**Porosity (Neutron, NMR, and Density):** These variables measure the amount of pore space between sediments. Neutron porosity is a measurement of the hydrogen index (like water or hydrate), and NMR porosity distinguishes between fluids in pore space and fluids that are bound to nearby clay.

**P-wave Velocity:** The p-wave velocity plot measures the speed of sound as it travels through the sediments. Gas hydrates make the sediment more solid, therefore causing an increase in seismic velocity. 

The most critical relationship observed is between high resistivity and high velocity. This correlation can be observed between 135-165 mbsf, where resistivity sharply increases while porosity measurements are very low. In this zone, non-conductive gas hydrate is stiffening the sediments and displacing the conductive pore water. At the same depth, the gamme ray plot suggest that hydrate accumulations are between the grains of coarser, larger-grained layers where the higher porosity allows for more hydrate to saturate. 

Overall, each of the variables interacts uniquely with the others. Increased porosity corresponds with coarser-grained sediments, shown by decreases in gamma radiation. Gas reservoirs tend to accrue in larger pore spaces, as shown by the velocity and resistivity plots. Density logs reveal the history of the area, including landslide masses. Under normal conditions, density increases and porosity decreases with depth due to overlying pressure. Abrupt changes in density or porosity, as seen at 70 mbsf, can therefore mark a landslide block that moved over the older, more compacted layers below. When framing the variables together, the data shows that the slow-moving landslides are likely influenced by gas hydrates accumulating in highly porous layers, which create zones of high pressure that facilitate slow movement on the seafloor.
